```python
import numpy as np
import matplotlib.pyplot as plt

# Generate non-stationary time series
t = np.linspace(0, 10, 100)
y = 2*t**3 - 3*t**2 + np.random.randn(100)

# Plot the time series
plt.plot(t, y)
plt.title('Non-stationary Time Series')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()

# First order differencing
y_diff1 = np.diff(y)

# Plot first differencing
plt.plot(t[1:], y_diff1)
plt.title('First Order Differencing')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()

# Second order differencing
y_diff2 = np.diff(y_diff1)

# Plot second differencing
plt.plot(t[2:], y_diff2)
plt.title('Second Order Differencing')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
```
